```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import math
import matplotlib.pyplot as plt

def logit(x):
    return 1 / (1+math.exp(x))

y = []
x = []
for k in range(-5, 5):
    y.append(logit(k))
    x.append(k)

plt.figure(figsize=(16,8))
plt.scatter(x=x,y=y,color='red')
```
